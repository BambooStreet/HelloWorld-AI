### 서울시 노동포털

```python
import json

def create_chunk(portal, category, content):
    #chunk = f"포털: {portal}\n"
    chunk = f"category: {category}\n"
    #chunk += f"세부 카테고리: {subcategory}\n"
    chunk += f"text:\n{content}"
    return chunk

# JSON 파일 읽기
with open('data\\raw_data\\Web\\서울노동포털.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

updated_data = []

for portal, categories in data.items():
    for category, subcategories in categories.items():
        for subcategory, contents in subcategories.items():
            content = "\n".join(contents)
            item = {
                #"portal": portal,
                #"category": category,
                #"subcategory": subcategory,
                #"text": content,
                "scource" : portal,
                "title" : subcategory,
                "content": create_chunk(portal, category, content)
            }
            updated_data.append(item)

# 수정된 데이터를 JSON 형식으로 저장
with open('서울노동포털_contents.json', 'w', encoding='utf-8') as f:
    json.dump(updated_data, f, ensure_ascii=False, indent=2)

print(f"총 {len(updated_data)}개의 항목이 서울노동포털_contents.json 파일로 저장되었습니다.")

# 첫 번째 항목 출력 (예시)
print("\n첫 번째 항목:")
print(json.dumps(updated_data[0], ensure_ascii=False, indent=2))
```

### 경기도 외국인 상담 센터

```python
import json

def create_chunk(item):
    #chunk = f"제목: {item['제목']}\n"
    chunk = f"작성일: {item['작성일']}\n"
    chunk += f"상담지원단체: {item['상담지원단체']}\n"
    chunk += f"상담유형: {item['상담유형']}\n"
    chunk += f"거주지역: {item['거주지역']}\n"
    chunk += f"국적: {item['국적']}\n"
    chunk += f"체류자격: {item['체류자격']}\n"
    chunk += f"상담내용: {item['상담내용']}\n"
    chunk += f"진행 과정 및 결과: {item['진행 과정 및 결과']}\n"
    if item['관련법령 및 정보']:
        chunk += f"관련법령 및 정보: {item['관련법령 및 정보']}\n"
    if item['평가 및 의의']:
        chunk += f"평가 및 의의: {item['평가 및 의의']}\n"
    chunk += f"작성자: {item['작성자']}\n"
    chunk += f"URL: {item['url']}"
    return chunk

# 원본 데이터 파일 경로
input_file = r"data\raw_data\Web\경기도외국인지원센터_상담사례.json"
# 결과 저장 파일 경로
output_file = r"경기도외국인지원센터_상담사례_contents.json"

# JSON 파일 읽기
with open(input_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

updated_data = []

for item in data:
    content = create_chunk(item)
    updated_item = {
        "source": "경기도외국인지원센터_상담사례",
        "title" : item['제목'],
        "content" : content
    }
    updated_data.append(updated_item)

# JSON 형식으로 저장
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(updated_data, f, ensure_ascii=False, indent=2)

print(f"총 {len(data)}개의 항목이 {output_file} 파일로 저장되었습니다.")

# 첫 번째 항목 출력 (예시)
print("\n첫 번째 항목:")
print(json.dumps(updated_data[0:2], ensure_ascii=False, indent=2))
```

### 키워드 처리 필요 데이터

```python
import json

# 처리할 파일 목록
files = [
    'easylaw_kr_contents',
    #'국내법률_민형사_근로기준법_판결문_contents',
    'legalqa_contents',
    #'법률규정텍스트_근로자_contents',
    '대한법률구조공단_상담사례_contents',
    '서울노동포털_contents',
]

# 결과를 저장할 리스트
result = []

# 각 파일을 처리
for i, file_name in enumerate(files):
    print(file_name)
    try:
        # 파일 읽기
        with open(f'data\contents\{file_name}.json', 'r', encoding='utf-8') as file:
            data = json.load(file)
            print(type(data))
        
        # 데이터가 리스트인 경우 처리
        if isinstance(data, list):
            for item in data:
                if 'content' in item:
                    result.append(item)

        else:
            # 데이터가 딕셔너리인 경우 (예외 처리)
            if 'content' in data:
                result.append(item)
    
    except FileNotFoundError:
        print(f"파일을 찾을 수 없습니다: {file_name}.json")
    except json.JSONDecodeError:
        print(f"JSON 디코딩 오류: {file_name}.json")
    except Exception as e:
        print(f"오류 발생: {file_name}.json - {str(e)}")

# 결과를 새 JSON 파일로 저장
with open('전처리필요.json', 'w', encoding='utf-8') as outfile:
    json.dump(result, outfile, ensure_ascii=False, indent=2)

print(f"병합된 {len(result)}개의 JSON 파일이 생성되었습니다: 전처리필요.json")
```

### 키워드별로 거르기

```python
import json

# 키워드 리스트 정의
keywords = ["외국인", "근로", "노동", "임금", "휴가", "비자", "직장", "월급"]

with open('data\contents\전처리필요.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 키워드를 포함하는 항목만 필터링
filtered_data = [item for item in data if any(keyword in item['content'] for keyword in keywords)]

# 필터링된 결과를 JSON 파일로 저장
with open('전처리완료.json', 'w', encoding='utf-8') as file:
    json.dump(filtered_data, file, ensure_ascii=False, indent=2)

print(f"필터링된 {len(filtered_data)}개의 데이터가 '전처리완료.json' 파일로 저장되었습니다.")
```

다른 데이터랑 결합

```python
import json
import os

def merge_json_files(directory):
    merged_data = []

    # 디렉토리 내의 모든 JSON 파일을 순회
    for filename in os.listdir(directory):
        
        if filename.endswith('.json'):
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                try:
                    data = json.load(file)
                    # 각 파일의 데이터가 리스트인지 확인
                    if isinstance(data, list):
                        for item in data:
                            if all(key in item for key in ["source", "title", "content"]):
                                merged_data.append(item)
                    elif isinstance(data, dict):
                        if all(key in data for key in ["source", "title", "content"]):
                            merged_data.append(data)
                except json.JSONDecodeError:
                    print(f"Error decoding JSON from file: {filename}")

    return merged_data

# JSON 파일들이 있는 디렉토리 경로
directory = 'data/contents/'

# JSON 파일들 병합
merged_data = merge_json_files(directory)

# 병합된 데이터를 새 JSON 파일로 저장
output_file = 'merged.json'
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(merged_data, f, ensure_ascii=False, indent=2)

print(f"총 {len(merged_data)}개의 항목이 {output_file} 파일로 저장되었습니다.")
```
